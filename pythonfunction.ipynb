```python
def even_odd(num):
  if(num%2==0):
    print(num,"is even")
  else:
    print(num,"is odd")

even_odd(5)
```

```python
def calculate(a,b):
  print("sum is",a+b)
  print("diff is",a-b)
  print("mul is",a*b)
calculate(10,5)
```

```python
def factorial(num):
  fact=1
  for i in range(1,num+1):
    fact=fact*i

  print(fact)
factorial(5)
```
